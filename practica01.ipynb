```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
data = pd.read_csv('AirPassengers.csv', index_col='Month', parse_dates=True)
data.head()
```

```python
data.plot()
plt.show()
```

```python
model = ARIMA(data['#Passengers'], order=(1, 1, 1))
fitted_model = model.fit()
print(fitted_model.summary())
```

```python
model = ARIMA(data['#Passengers'], order=(2, 1, 2))
fitted_model = model.fit()
print(fitted_model.summary())
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Diferenciar la serie para que sea estacionaria, si es necesario
data_diff = data['#Passengers'].diff().dropna()

# Graficar ACF y PACF
fig, axes = plt.subplots(1, 2, figsize=(15,5))
plot_acf(data_diff, ax=axes[0])
plot_pacf(data_diff, ax=axes[1])
plt.show()
```

```python
model1 = ARIMA(data['#Passengers'], order=(1, 1, 1))
fitted_model1 = model1.fit()
#print(fitted_model2.summary())
model2 = ARIMA(data['#Passengers'], order=(1, 1, 2))
fitted_model2 = model2.fit()
model3 = ARIMA(data['#Passengers'], order=(1, 1, 3))
fitted_model3 = model3.fit()
model4 = ARIMA(data['#Passengers'], order=(2, 1, 1))
fitted_model4 = model4.fit()
model5 = ARIMA(data['#Passengers'], order=(2, 2, 1))
fitted_model5 = model5.fit()
model6 = ARIMA(data['#Passengers'], order=(2, 2, 2))
fitted_model6 = model6.fit()
model7 = ARIMA(data['#Passengers'], order=(3, 1, 1))
fitted_model7 = model7.fit()
model8 = ARIMA(data['#Passengers'], order=(3, 2, 1))
fitted_model8 = model8.fit()
```

```python
print("ARIMA(1,1,1): \t LL =", fitted_model1.llf, "\t AIC = ", fitted_model1.aic)
print("ARIMA(1,1,2): \t LL =", fitted_model2.llf, "\t AIC = ", fitted_model2.aic)
print("ARIMA(1,1,3): \t LL =", fitted_model3.llf, "\t AIC = ", fitted_model3.aic)
print("ARIMA(2,1,1): \t LL =", fitted_model4.llf, "\t AIC = ", fitted_model4.aic)
print("ARIMA(2,2,1): \t LL =", fitted_model5.llf, "\t AIC = ", fitted_model5.aic)
print("ARIMA(2,2,2): \t LL =", fitted_model6.llf, "\t AIC = ", fitted_model6.aic)
print("ARIMA(3,1,1): \t LL =", fitted_model7.llf, "\t AIC = ", fitted_model7.aic)
print("ARIMA(3,2,1): \t LL =", fitted_model8.llf, "\t AIC = ", fitted_model8.aic)
```

Para seleccionar el mejor modelo ARIMA, se utiliza el criterio AIC (Criterio de Informacin de Akaike), que equilibra la precisin del modelo y su simplicidad, penalizando modelos excesivamente complejos. El objetivo es elegir el modelo con el AIC ms bajo, ya que esto sugiere un buen ajuste sin sobreajuste.

en los resultados, se presentan varios modelos ARIMA(1,1,1) con distintos valores de AIC. Al comparar estos valores, el modelo con el AIC ms bajo es 1378.34, correspondiente al modelo ARIMA(1,1,1). Esto indica que, entre las opciones mostradas, este es el modelo ms adecuado para los datos analizados, ya que ofrece un buen ajuste con menor complejidad en comparacin con los otros modelos evaluados.

En resumen, ARIMA(1,1,1) con AIC = 1378.34 es el modelo recomendado para esta serie temporal, de acuerdo con el criterio AI
